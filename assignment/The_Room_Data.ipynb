We shall sample the data and drop the columns that will not be used in the analysis of the data. 
Based on the data size we shall use *__Simple random sampling technique__*



```python
users_df.head()
```

```python
# we shall use 1800 records for our analysis
users_df_sam = users_df.sample(1800)
users_df_sam
```

```python
#looking at the null values
print(users_df_sam.apply(lambda x: sum(x.isnull())))

users_df_sam.columns.to_list()
```

```python
#dropping the columns that will not be useful in the analysis
users_df_sam.drop(['Languages','Customer Call Status','Pathfinder Assigned', 'Interviewed','Industry'
                   ,'City','Has Linkedin','Age','Citizenship'], axis = 1, inplace=True)
```

```python
# dealing with the missing values in the lifetime Days columns
# we shall use the mean lifetime days to get rid of the null values

users_df_sam["Lifetime Days"].fillna((users_df_sam['Lifetime Days'].mean()), inplace = True)
users_df_sam['Lifetime Days'].isnull().sum()
```

```python
print(users_df_sam.apply(lambda x: sum(x.isnull())))
```

```python
# from this size of 1800 we shall ommit the NaN 
users_df_sam.dropna(inplace=True)
```

```python
users_df_sam.shape
```

For our analysis we shall use the 1174 sample size, which contains no  missing values

```python
users_df_sam.groupby('Affiliation')['ID'].count()
```

```python
users_df_sam.head()
```

```python
# adding a new column inorder to have more complteness for the present state of the values
# we shall use the life time days to assign values in years 
users_df_sam['Yearly_count'] = " "
```

```python
# creating an empty list and dividing the number of days by 365 days
# to get the number in years, this will help in further analysis of the data
a = []
for i in users_df_sam['Lifetime Days']:
    i  = i /365
    i = round(i,1)
    a.append(i)  
 
```

```python
users_df_sam['Yearly_count'] = a
```

```python

users_df_sam.head()
```

### Exploratory Analysis

Looking through the data to extract useful information 

```python
fig = px.histogram(users_df_sam['Yearly_count'], color=users_df_sam['Membership Type'],
                  nbins = 10, hover_name=users_df_sam['Country'], marginal='rug',
                  height=800, width= 1000, title = "Distribution of members based on the years")
fig.update_xaxes(
ticksuffix = "yrs", showgrid = True)
fig.update_layout ( # customize the font and the legend
 font_family = "Rockwell", 
legend = dict (
title = "Sample distribution", orientation = "h"))
fig.show()
```

### Observations
Quick take aways from the historgram above is that we have a majority of the members are within 6 months range and they are of free trial (621). The members are mostly from Kenya, USA and Switzerland, and premium paids lie within this region but only account for 149 members
