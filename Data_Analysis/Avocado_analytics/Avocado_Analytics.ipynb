```python
#importing the neccesary libraries
import dash #helps intitlialize the application 
import dash_core_components as dcc #to create the interactive components
import dash_html_components as html # helps access the HTML tags
import pandas as pd
```

```python
#importing the data
data = pd.read_csv('https://raw.githubusercontent.com/realpython/materials/master/python-dash/additional_files/avocado.csv')
```

```python
data['Date'] = pd.to_datetime(data['Date'], format = "%Y-%m-%d")
data.sort_values('Date', inplace= True)
```

```python
data.head()
```

```python
data = data.query("type == 'conventional' and region == 'Albany'")
data.head()
```

```python
app = dash.Dash(__name__) #creating an instance of the Dash class
```

## Defining the Layout of the Dash Application

This will dictate the overall look of the app. It will be having two graphs

### Styling the Dash Application 

* We shall use style arguments to customaize the dashboard, it takes Python dictionary Key bvalue pairs consisting of the names of CSS properties and the values to be set 

```python
#Customizing the Styles of Components
# We shall use the ClassName arguments in Dash components
# The className adds a corresponding class selector to each component that
#compose the header of the dashboard
external_stylesheets = [
    {
        "href": "https://fonts.googleapis.com/css2?"
        "family=Lato:wght@400;700&display=swap",
        "rel": "stylesheet",
    },
]
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
app.title = "Avocado Analytics: Understand Your Avocados!"

app.layout = html.Div(
    children=[
        html.Div(
            children=[
                html.P(children="🥑", className="header-emoji"),
                html.H1(
                    children="Avocado Analytics", className="header-title"
                ),
                html.P(
                    children="Analyze the behavior of Avocado prices"
                    " and the number of avocados sold in the US"
                    " between 2015 and 2018",
                    className="header-description",
                ),
            ],
            className="header",
        ),
        html.Div(
            children=[
                html.Div(
                    children=dcc.Graph(
                        id="price-chart",
                        config={"displayModeBar": False},
                        figure={
                            "data": [
                                {
                                    "x": data["Date"],
                                    "y": data["AveragePrice"],
                                    "type": "lines",
                                    "hovertemplate": "$%{y:.2f}"
                                                     "<extra></extra>",
                                },
                            ],
                            "layout": {
                                "title": {
                                    "text": "Average Price of Avocados",
                                    "x": 0.05,
                                    "xanchor": "left",
                                },
                                "xaxis": {"fixedrange": True},
                                "yaxis": {
                                    "tickprefix": "$",
                                    "fixedrange": True,
                                },
                                "colorway": ["#17B897"],
                            },
                        },
                    ),
                    className="card",
                ),
                html.Div(
                    children=dcc.Graph(
                        id="volume-chart",
                        config={"displayModeBar": False},
                        figure={
                            "data": [
                                {
                                    "x": data["Date"],
                                    "y": data["Total Volume"],
                                    "type": "lines",
                                },
                            ],
                            "layout": {
                                "title": {
                                    "text": "Avocados Sold",
                                    "x": 0.05,
                                    "xanchor": "left",
                                },
                                "xaxis": {"fixedrange": True},
                                "yaxis": {"fixedrange": True},
                                "colorway": ["#E12D39"],
                            },
                        },
                    ),
                    className="card",
                ),
            ],
            className="wrapper",
        ),
    ]
)

```

```python
if __name__ == "__main__":
    app.run_server(debug = True)
```
